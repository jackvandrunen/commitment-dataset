```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('new_model_1.json') as f:
    results = np.array(json.load(f))
    empiricalSuccess = results[:,0]
    commitProb = results[:,1]
    successGivenCommit = results[:,2]
    successGivenNotCommit = results[:,3]


print('P(Commit) > 0.9 =', np.sum(commitProb > 0.9))
print('P(Success | P(Commit) > 0.9) =', np.mean((commitProb*successGivenCommit + (1-commitProb)*successGivenNotCommit)[np.where(commitProb > 0.9)]))
print('P(Commit) < 0.1 =', np.sum(commitProb < 0.1))
print('P(Success | P(Commit) < 0.1) =', np.mean((commitProb*successGivenCommit + (1-commitProb)*successGivenNotCommit)[np.where(commitProb < 0.1)]))

successProb = (commitProb*successGivenCommit + (1-commitProb)*successGivenNotCommit)
print(np.sum(successProb > 0.995))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 3.5))

ax[0].plot([0, 1], [0, 1], 'k--')
ax[0].plot(np.sort(commitProb), np.arange(1, 1001) / 1000, '.')
ax[0].set_xlabel('Commitment Probability')
ax[0].set_ylabel('Proportion of Simulations')
ax[0].label_outer()

ax[1].plot(np.sort(successGivenCommit), np.arange(1, 1001) / 1000, '.', label='P(Success | Commit)')
ax[1].plot(np.sort(successGivenNotCommit), np.arange(1, 1001) / 1000, '.', label='P(Success | ~Commit')
ax[1].set_xlabel('Success Probability')
ax[1].set_ylabel('Proportion of Simulations')
ax[1].legend()
ax[1].label_outer()

plt.suptitle('Model 1 Results')
plt.savefig('figure_1.pdf')
plt.show()
```
